```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Superstore.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.sort_values('Sales',ascending=False).head(10)
```

```python
data.sort_values('Profit',ascending=False).head(10)
```

```python
data.drop(['Row ID','Postal Code'],axis=1).describe()
```

```python
corr_data=data.drop(['Row ID'],axis=1).corr(method='pearson')
sns.heatmap(corr_data,annot=True,cmap='inferno')
```

```python
sns.regplot(data=data,x='Sales',y='Profit')
```

```python
sns.regplot(data=data,x='Discount',y='Profit',color='b')
```

```python
data['Year']=pd.DatetimeIndex(data['Ship Date']).year
data.head()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.lineplot(data=data,x='Year',y='Sales',ci=None)
plt.xticks(rotation=90)
plt.subplot(1,2,2)
sns.lineplot(data=data,x='Year',y='Profit',ci=None)
plt.xticks(rotation=90)
sns.set_style('whitegrid')
plt.suptitle('Sales v/s Profit')
```

```python
sales_subcategory=data.groupby('Sub-Category')['Sales'].sum()
sales_subcategory
```

```python
plt.figure(figsize=(16,12))
plt.pie(sales_subcategory,labels=sales_subcategory.index,autopct='%.1f%%')
plt.ylabel('')
```

```python
sns.barplot(x=sales_subcategory.index,y=sales_subcategory)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=data,x='Region',y='Sales',hue='Year',ci=None)
plt.legend(loc='upper right')
```

```python
top_cities=data.sort_values('Sales',ascending=False).head(10)
s_t=top_cities.groupby('City')['Sales'].sum()
s_t
```

```python
bottom_cities=data.sort_values('Sales',ascending=True).head(12)
s_b=bottom_cities.groupby('City')['Sales'].sum()
s_b
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.barplot(x=s_t.index,y=s_t)
plt.xticks(rotation=90)
plt.subplot(1,2,2)
sns.barplot(x=s_b.index,y=s_b)
plt.xticks(rotation=90)
```
